# Learning to flip edges in triangular meshes

```julia
using TriMeshGame
using MeshPlotter
using ProximalPolicyOptimization
using Distributions: Categorical
using Statistics: mean, std

TM = TriMeshGame
MP = MeshPlotter
PPO = ProximalPolicyOptimization
```

# Setting up the game environment

First let's write a function that will generate the initial mesh,

```julia
function make_test_mesh()
    vertices = [0.0  1.0  0.5  0.5
                0.0  0.0  0.5 -0.5]
    connectivity = [1 1
                    2 4
                    3 2]
    mesh = TM.Mesh(vertices, connectivity)
    return mesh
end
```

Let's visualize the initial mesh,

```julia
mesh = make_test_mesh()
MP.plot_mesh(TM.active_vertex_coordinates(mesh), 
    TM.active_triangle_connectivity(mesh), number_vertices = true, number_elements = true,
figsize = 8);
```

Our game is to decide whether to flip the included edge or not. We do this to try to attain the desired degree for each vertex. The desired degree is specified by a vector of integers. For example desired_degree = [3,3,2,2] says that we would like the degree of vertex 1 to be 3 and so on.

```julia
function get_desired_degree()
   if rand() < 0.5
        return [4,4,1,1]
    else
        return [2,2,3,3]
    end
end
```

Notice that with the above definition of desired degree, with 50% probability, we would like to flip the edge.

Now we will define our game environment. This will help us keep track of how many actions we have taken, the reward for each action, and whether or not the environment has terminated.

```julia
function make_trimesh_env(desired_degree)
    mesh = make_test_mesh()
    max_actions = 1
    env = TM.GameEnv(mesh, desired_degree, max_actions)
    return env
end
```

We define a wrapper to the `TriMeshGame.GameEnv` so that we can reset it easily.

```julia
mutable struct TwoElementEnv
    env
    function TwoElementEnv(;desired_degree = get_desired_degree())
        env = make_trimesh_env(desired_degree)
        return new(env)
    end
end
```

We also overload the `PPO.reset!` function for our environment,

```julia
function PPO.reset!(wrapper::TwoElementEnv; desired_degree = get_desired_degree())
   env = make_trimesh_env(desired_degree)
    wrapper.env = env
end
```

We can use `MeshPlotter` to plot the environment with vertex scores,

```julia
function plot_env(env)
    v = TM.active_vertex_coordinates(env.mesh)
    c = TM.active_triangle_connectivity(env.mesh)
    vs = TM.active_vertex_score(env)
    fig = MP.plot_mesh(v, c, vertex_score = vs, vertex_size = 30, figsize = 8)
    return fig
end
```

```julia
wrapper = TwoElementEnv()
plot_env(wrapper.env);
```

```julia
PPO.reset!(wrapper)
plot_env(wrapper.env);
```

Run the above code cell a few times to see that the vertex scores change. There are two states that come from the random initialization of desired degree. When the vertex scores are `[+1,+1,-1,-1]`, flipping the edge gives a reward of +4. In the other state, flipping the edge gives a reward of -4. So it is desirable to flip the edge in only one of the states with probability 50%.

# Defining the `state` of the game environment

How can we decide whether or not to flip the edge? If we are given the vertex scores of the vertices at the end-points of the edge, and the vertex scores of the vertices opposite the edge, we can determine if flipping the edge is desirable. But there is an ambiguity in the ordering of each pair of these numbers -- for example the order of the vertices at the end-points of the edges may be reversed. This is because edges are not oriented. However, half-edges _are_ oriented. Each half-edge has a unique start and end-point induced by the orientation. The vertices opposite the half-edge may also be ordered by whether the vertex belongs to the same triangle as the current half-edge or if it belongs to the neighboring triangle. This ends up doubling the number of actions since there are ~ twice as many half-edges as edges. This is not necessarily a bad thing since an agent playing this game will inherently learn the symmetries involved -- it's a kind of data augmentation.

We will write a function to create a template around each half edge. The template is an ordered list of vertices whose vertex scores can be used to determine if an edge should be flipped or not. `TriMeshGame` provides several utilities for this,

```julia
function get_pairs(data, pairs)
    return TM.zero_pad(data)[:, pairs]
end

function make_template(mesh)
    # pairs is a vector of half-edge indices. pairs[i] is the global index of the half-edge that is the twin
    # edge of global half-edge i
    pairs = TM.make_edge_pairs(mesh)
    
    # The flattened connectivity is the vertex opposite each half-edge
    tri_vertices = reshape(mesh.connectivity, 1, :)
    
    # cycle_edges gathers information from neighboring half-edges in the same triangle
    ct = TM.cycle_edges(tri_vertices)
    
    # getting the `pairs` columns effectively gathers information from twin edges.
    # we zero_pad for half-edges on the boundary which don't have a twin and default to a zero value.
    p = get_pairs(tri_vertices, pairs)
    
    # Concatenate these two matrices and return
    template = vcat(ct, p)

    return template
end
```

For the simple two element mesh we are considering here, the above definition is sufficient. However, in more complex meshes, it may be beneficial to iterate the `cycle_edges` and `get_pairs` operations since this leads to a deeper stencil around each half-edge. We will explore this in other examples.

```julia
mesh = make_test_mesh()
template = make_template(mesh)
```

```julia
MP.plot_mesh(TM.active_vertex_coordinates(mesh), 
    TM.active_triangle_connectivity(mesh), number_vertices = true, number_elements = true,
figsize = 8, internal_order = true);
```

Observe the vertex indices in the above figure. As an example, consider half-edge 2 in triangle 2. This half-edge has global index 5. This is the directed edge from vertex 2 to vertex 1 in triangle 2 (opposite the red 2 in triangle 2). Observe that column 5 in the template is `[4,2,1,3]`. This is the ordered list of vertices surrounding half-edge 5. This half-edge's twin is the half-edge with global index 3 -- the directed edge from vertex 1 to vertex 2 in triangle 1 (opposite the red vertex 3). Observe that column 3 in the template is `[3,1,2,4]` which is a re-arrangement of the vertices in column 5.

In general, the vertices associated with each half-edge will be `[opposite_vertex_from_same_triangle, source_vertex, destination_vertex, opposite_vertex_from_neighbor_triangle]`. Using the vertex scores of the vertices in the template, it is possible to decide if a particular edge should be flipped. For example, if the scores of the vertices in the template are `[-1,+1,+1,-1]` then we would want to flip the edge.

Not all actions are permitted. Since `TriMeshGame` allocates a buffer of triangles, there are many inactive triangles which should not be selected. To deal with this, we create an `action_mask`.

```julia
function make_action_mask(active_triangle; actions_per_edge=2)
    actions_per_triangle = 3 * actions_per_edge
    requires_mask = repeat(.!active_triangle', inner=(actions_per_triangle, 1))
    mask = vec([r ? -Inf32 : 0.0f0 for r in requires_mask])
    return mask
end
```

```julia
action_mask = make_action_mask(mesh.active_triangle)
```

Notice that `action_mask` takes on a value of `-Inf` for actions associated with inactive triangles. The reason for this will become clear shortly. We can now define the state of the game, and define a `struct` to store the required information,

```julia
struct StateData
    vertex_score::Any
    action_mask
end

function Base.show(io::IO, s::StateData)
    num_actions = length(s.action_mask)
    println(io, "StateData")
    println(io, "\t $num_actions actions")
end

# missing_val is used for boundary half-edges which don't have associated twin-edges.
function val_or_missing(vector, template, missing_val)
    return [t == 0 ? missing_val : vector[t] for t in template]
end

function PPO.state(wrapper)
    env = wrapper.env
    mesh = env.mesh
    
    template = make_template(mesh)
    vs = val_or_missing(env.vertex_score, template, +10)

    am = make_action_mask(mesh.active_triangle)

    s = StateData(vs, am)

    return s
end
```

```julia
s = PPO.state(wrapper)
```

# Computing a `policy` over actions

Since `TriMeshGame` already supplies an ordered list of vertices for each half-edge, it is sufficient for our `policy` to be quite simple. In fact, for this simple game environment, all you need is a linear layer. The input to the network is 4 numbers, and the output is the logit of selecting a given action. In our game each half-edge is associated with an action -- flip or do nothing. Flipping gives you a reward equal to the change in the environment score. Doing nothing gives you a reward of zero. Based on how we set up the game environment, we would want to do nothing about 50% of the time.

```julia
using Flux
```

```julia
policy = Dense(4,2)
```

We now overload `PPO.action_probabilities`,

```julia
function PPO.action_probabilities(policy, state)
    vertex_score, action_mask = state.vertex_score, state.action_mask
    
    # adding the action mask before running softmax ensures that invalid actions are assigned a probability of
    # zero
    logits = vec(policy(vertex_score)) + action_mask
    p = softmax(logits)
    return p
end
```

```julia
ap = PPO.action_probabilities(policy, s)
```

Notice that the actions associated with invalid triangles have a probability of zero. Therefore, they will not be selected by our policy.

# `step`ping the environment

We need to define what it means to take an action in our environment.

```julia
function step_env!(wrapper, triangle_index, half_edge_index, action_type, no_action_reward = -4)
    env = wrapper.env

    @assert TM.is_active_triangle(env.mesh, triangle_index) "Attempting to act on inactive triangle $triangle_index with action ($triangle_index, $half_edge_index, $action_type)"
    @assert action_type in (1, 2) "Expected action type in {1,2} got type = $action_type"
    @assert half_edge_index in (1, 2, 3) "Expected edge in {1,2,3} got edge = $half_edge_index"
    @assert !env.is_terminated
    
    if action_type == 1
        TM.step_flip!(env, triangle_index, half_edge_index, no_action_reward = no_action_reward)
    else
        TM.step_nothing!(env, reward = 0)
    end
end

# since our actions are defined by a linear index, we need to convert the index to an action
function index_to_action(index; actions_per_edge=2)
    actions_per_triangle = 3 * actions_per_edge

    triangle_index = div(index - 1, actions_per_triangle) + 1

    triangle_action_idx = rem(index - 1, actions_per_triangle)
    edge = div(triangle_action_idx, actions_per_edge) + 1
    action = rem(triangle_action_idx, actions_per_edge) + 1

    return triangle_index, edge, action
end

function PPO.step!(wrapper, action_index; no_action_reward = -4)

    triangle_index, half_edge_index, action_type = index_to_action(action_index)
    step_env!(wrapper, triangle_index, half_edge_index, action_type)
end
```

# Rewards and terminal states

We overload the functions `PPO.reward` and `PPO.is_terminal`,

```julia
function PPO.reward(wrapper)
    env = wrapper.env
    return env.reward
end

function PPO.is_terminal(wrapper)
    env = wrapper.env
    return env.is_terminated
end
```

# PPO algorithm step 1: collecting rollouts

In the first step of the PPO algorithm, we collect multiple rollouts of our current policy on the environment. To do this, `PPO` provides `EpisodeData` which stores the information required to perform optimization. This includes states, selected actions, their probabilities, rewards, and if the state was terminal.

Here's how we can collect 100 rollouts. Each rollout consists of a full episode until the environment terminates.

```julia
const episodes_per_rollout = 100
rollouts = PPO.EpisodeData()
PPO.collect_rollouts!(rollouts, wrapper, policy, episodes_per_rollout)
```

```julia
rollouts
```

```julia
rollouts.rewards
```

Most of the rewards are -4 because our agent is currently acting randomly. This will change after training. In order to train on the collected data, we first need to compute the value of each state. This is equal to the discounted return from that state. (We currently don't implement baselines.) Since the current game has only one step, this doesn't matter, but let's do it anyway.

```julia
const discount = 1.0
PPO.compute_state_value!(rollouts, discount)
```

# PPO algorithm step 2: Optimization

In step 2 of the PPO algorithm, we use the collected rollouts to train our policy to pick better actions. There are many steps involved in this. In each epoch,
- Shuffle the rollouts
- Chunk them into mini-batches
- Perform gradient descent on each mini-batch using the PPO loss function

We need to define a few things for `PPO` before we can train. Firstly, we need to specify how to __batch__ our `state`. This simply means concatenating each mini-batch of states.

```julia
function PPO.batch_state(state_data_vector)
    vs = [s.vertex_score for s in state_data_vector]
    am = [s.action_mask for s in state_data_vector]
    
    batch_vertex_score = cat(vs..., dims = 3)
    batch_action_mask = cat(am..., dims = 2)
    return StateData(batch_vertex_score, batch_action_mask)
end
```

We also need to define how to evaluate our policy on a batch of states.

```julia
function PPO.batch_action_probabilities(policy, state)
    vertex_score, action_mask = state.vertex_score, state.action_mask
    num_features, num_actions, num_batch = size(vertex_score)
    logits = reshape(policy(vertex_score), :, num_batch) + action_mask
    probs = softmax(logits, dims=1)
    return probs
end
```

Now we can train our policy,

```julia
const batch_size = 10
const num_epochs = 20
const epsilon = 0.05 # this is the epsilon controlling the "proximal" part of the PPO algorithm
```

```julia
optimizer = Adam(1e-4)
```

```julia
PPO.ppo_train!(policy, optimizer, rollouts, epsilon, batch_size, num_epochs)
```

# Evaluating the `policy`

The full PPO algorithm iterates multiple times between step 1 and step 2 above. We have a convenience function `ppo_iterate` that will do this for us. It is helpful to evaluate the `policy` as we optimize to see how well we are doing. We can evaluate our policy as follows,

```julia
function single_trajectory_return(policy, wrapper)
    env = wrapper.env
    ret = 0
    done = PPO.is_terminal(wrapper)

    while !done
        probs = PPO.action_probabilities(policy, PPO.state(wrapper))
        action = rand(Categorical(probs))

        PPO.step!(wrapper, action)
        done = PPO.is_terminal(wrapper)
        ret += PPO.reward(wrapper)
    end
    return ret
end

function average_returns(policy, wrapper, num_trajectories)
    ret = zeros(num_trajectories)
    for idx = 1:num_trajectories
        PPO.reset!(wrapper)
        ret[idx] = single_trajectory_return(policy, wrapper)
    end
    return mean(ret), std(ret)
end
```

```julia
function evaluator(policy, wrapper; num_trajectories = 100)
    ret, dev = average_returns(policy, wrapper, num_trajectories)
    return ret, dev
end
```

```julia
evaluator(policy, wrapper)
```

```julia
const num_iter = 1000
```

```julia
returns, deviation = PPO.ppo_iterate!(policy, wrapper, optimizer, episodes_per_rollout, discount, 
epsilon, batch_size, num_epochs, num_iter, evaluator)
```

```julia
using PyPlot
```

```julia
plot(returns)
```

Observe that the returns plateau at 2. In our game environment, with 50% probability we can get a reward of 4 and with 50% probability we can get a reward of 0. So the expected reward is about 2 which is what our model converges to.
